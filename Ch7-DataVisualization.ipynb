```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
```

```python
plt.style.use('seaborn')
```

```python
mpl.rcParams['font.family'] = 'serif'
```

```python
%matplotlib inline
```

# 1D dataset

```python
#setting a limit
np.random.seed(1000) 
```

```python
#creating x and y axis
y = np.random.standard_normal(20)
x = np.arange(len(y))
```

```python
plt.plot(x,y)
```

```python
#manipulating grids and labels
plt.plot(y.cumsum())
plt.grid(False)
plt.axis('equal')
```

```python
#setting limits
plt.plot(y.cumsum())
plt.xlim(-1, 20)
plt.ylim(np.min(y.cumsum()) - 1,
        np.max(y.cumsum()) + 1)
```

```python
#modifying plot presentation
plt.figure(figsize=(10,6))
plt.plot(y.cumsum(), 'b', lw=1.5, linestyle='dashdot')
plt.plot(y.cumsum(), 'ro')
plt.xlabel('index')
plt.ylabel('value')
plt.title('Sample plot')
```

# 2D Dataset

```python
#Generating a 2-dimensional normal standard data set with 20 x 2 shape with cumsum() function to calculate cumulative sum
y = np.random.standard_normal((20, 2)).cumsum(axis=0)
print(y)
```

```python
#creating the 2D plot
plt.figure(figsize=(20,6))
plt.plot(y[:, 0], lw=1.5, label='first')
plt.plot(y[:, 1], lw=1.5, label='second')
plt.plot(y, 'ro')
plt.legend(loc=0)
plt.xlabel('index')
plt.ylabel('value')
plt.title('Simple Plot')
```

```python
#Recreating the 2D plot but 1st column is multipled by 100
y[:, 0] = y[:, 0] * 100
plt.figure(figsize=(20,6))
plt.plot(y[:, 0], lw=1.5, label='first')
plt.plot(y[:, 1], lw=1.5, label='second')
plt.plot(y, 'ro')
plt.legend(loc=0)
plt.xlabel('index')
plt.ylabel('value')
plt.title('Simple Plot')
```

```python
#Instead of changing the data via scaling, one approach is to introduce a 2nd y-axis
#y[:, 0] = y[:, 0] * 100
fig, ax1 = plt.subplots()
plt.plot(y[:, 0], color='b', lw=1.5, label='first')
plt.plot(y[:, 0], 'ro')
plt.legend(loc=4)
plt.xlabel('index')
plt.ylabel('value')
ax2 = ax1.twinx()
plt.plot(y[:, 1], color='g', lw=1.5, label='second')
plt.plot(y[:, 1], 'ro')
plt.legend(loc=1)
plt.xlabel('index')
plt.ylabel('value')
plt.title('Simple Plot')
```

```python
#2nd approach is using 2 subplots
plt.figure(figsize=(10,6))
plt.subplot(2,1,1) #(nrow,ncol,fignum)
plt.plot(y[:, 0], color='b', lw=1.5, label='first')
plt.plot(y[:, 0], 'ro')
plt.xlabel('index')
plt.ylabel('values')
plt.title('Simple Plot')
plt.subplot(2,1,2)
plt.plot(y[:, 1], color='g', lw=1.5, label='first')
plt.plot(y[:, 1], 'ro')
plt.xlabel('index')
plt.ylabel('values')
```

```python
#Creating a different plot style for the 2nd set
plt.figure(figsize=(10,6))
plt.subplot(1,2,1) #(nrow,ncol,fignum)
plt.plot(y[:, 0], color='b', lw=1.5, label='first')
plt.plot(y[:, 0], 'ro')
plt.xlabel('index')
plt.ylabel('values')
plt.title('Simple Plot')
plt.subplot(1,2,2)
plt.bar(np.arange(len(y)), y[:, 1], width=.5, label='second')
plt.xlabel('index')
plt.ylabel('values')
plt.title('Bar plot')
```

# Different plot styles

```python
#Scatter plot
y = np.random.standard_normal((1000, 2))
c = np.random.randint(0, 10, len(y))
plt.scatter(y[:, 0], y[:, 1], c=c, cmap='coolwarm')
plt.colorbar()
plt.xlabel('index')
plt.ylabel('values')
plt.title('Scatter plot')
```

```python
plt.hist(y, bins='auto')
plt.xlabel('index')
plt.ylabel('values')
plt.title('Histogram')
```

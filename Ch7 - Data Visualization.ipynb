```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
```

```python
plt.style.use('seaborn')
```

```python
mpl.rcParams['font.family'] = 'serif'
```

```python
%matplotlib inline
```

# 1D dataset

```python
#setting a limit
np.random.seed(1000) 
```

```python
#creating x and y axis
y = np.random.standard_normal(20)
x = np.arange(len(y))
```

```python
plt.plot(x,y)
```

```python
#manipulating grids and labels
plt.plot(y.cumsum())
plt.grid(False)
plt.axis('equal')
```

```python
#setting limits
plt.plot(y.cumsum())
plt.xlim(-1, 20)
plt.ylim(np.min(y.cumsum()) - 1,
        np.max(y.cumsum()) + 1)
```

```python
#modifying plot presentation
plt.figure(figsize=(10,6))
plt.plot(y.cumsum(), 'b', lw=1.5, linestyle='dashdot')
plt.plot(y.cumsum(), 'ro')
plt.xlabel('index')
plt.ylabel('value')
plt.title('Sample plot')
```

# 2D Dataset

```python
#Generating a 2-dimensional normal standard data set with 20 x 2 shape with cumsum() function to calculate cumulative sum
y = np.random.standard_normal((20, 2)).cumsum(axis=0)
print(y)
```

```python
#creating the 2D plot
plt.figure(figsize=(20,6))
plt.plot(y[:, 0], lw=1.5, label='first')
plt.plot(y[:, 1], lw=1.5, label='second')
plt.plot(y, 'ro')
plt.legend(loc=0)
plt.xlabel('index')
plt.ylabel('value')
plt.title('Simple Plot')
```
